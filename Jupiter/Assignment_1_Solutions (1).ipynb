Aşağıdaki varsayımlar doğrultusunda, Bölüm A'da yeterli miktarda para biriktirmeniz için kaç ay süreceğini belirleyeceğiz:


* Hayalini kurduğunuz evin maliyetini total\_cost olarak adlandırın.
* Peşinat için gereken maliyeti portion\_down\_payment olarak adlandırın. Basitlik için portion\_down\_payment = 0.25 (yüzde 25) varsayalım.
* Şu ana kadar biriktirdiğiniz tutarı current\_savings olarak adlandırın. Birikime şu an için 0 lirayla başlıyorsunuz.

* Yıllık maaşınız annual\_salary olsun.
* Peşinat için her ay biriktireceğiniz maaşınızın belirli bir miktarını ayıracağınızı varsayın. Bunu portion\_saved olarak adlandıralım. Bu değişken ondalık formda olmalıdır (örneğin, yüzde 10 için 0.1).

Her ayın sonunda, birikimleriniz yatırım getirisiyle artırılacak ve aylık maaşınızın yüzdesi eklenerek güncellenecektir (yıllık maaş / 12).

Ev için yeterli miktarda para biriktirmeniz için kaç ay süreceğini hesaplayan bir program yazın. Ana değişkenlerinizin ondalık sayılar olmasını isteyeceğiniz için, kullanıcı girişlerini ondalık sayılara dönüştürmelisiniz.

Programınız kullanıcıdan aşağıdaki değişkenleri girmesini istemelidir:

* Başlangıç yıllık maaşınız (annual\_salary)
* Birikim yapmak için maaşınızın yüzdesi (portion\_saved)
* Hayalini kurduğunuz evin maliyeti (total\_cost)

#### Test 1 

Yıllık maaşınızı girin: 120000

#### Test 2

### Solutiton

```python
annual_salary = int(input("Yıllık maaşınızı girin:   "))
portion_saved = float(input("Biriktirmek için maaşınızın yüzdesini ondalık olarak girin:   "))
total_cost = int(input("Hayalini kurduğunuz evin maliyetini girin:   "))

portion_down_payment = 0.25 * total_cost
current_savings = 0
r = 0.04 
number_of_months = 0

while current_savings < portion_down_payment:
    current_savings += portion_saved * annual_salary / 12 + current_savings * r / 12
    number_of_months += 1
    
print("Number of months:  " , number_of_months)
```

## Normal Distribution (Gaussian Distribution)

### Probability Density Function of the Normal Distribution

- unimodal and symmetric
- many algorithms in machine learning & statistics have normality assumptions
- two parameters: mean (center of the peak) and standard deviation (spread); $N(\mu, \sigma)$
- we can estimate parameters of $N(\mu, \sigma^2)$ by sample mean ($\bar{x})$ and sample variance ($s^2$)
- univariate Normal distribution:

$$f(x \mid \mu, \sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} \; \exp\bigg(-\frac{(x-\mu)^2}{2\sigma^2}\bigg)$$


- standard normal distribution with zero mean and unit variance, $N(0, 1)$:
